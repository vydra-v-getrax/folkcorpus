```python
# !pip install python-docx
from docx import Document
import re
import os
```

```python
# all_docs = []
# broken_docs = []
folder = "фольклорный архив"
for filepath in os.listdir(folder):
    if filepath.endswith('.docx'):
        print("Processing file: ", filepath)
        with open(os.path.join(folder, filepath), 'rb') as f:
            document = Document(f)
            
            tables = document.tables
            for table in tables:
                doc = {}
                
                if len(table.columns) != 4:
                    print('Wrong table. Number of columns: ', len(table.columns), filepath)
                    broken_docs.append(filepath)
                if len(table.rows) != 3:
                    print('Wrong table. Number of rows: ', len(table.rows), filepath)
                    broken_docs.append(filepath)

                else:
                    doc['Filename'] = os.path.join(folder, filepath)
                    doc['Number'] = table.rows[1].cells[0].text.strip()
                    doc['Genre'] = table.rows[1].cells[1].text.strip()
                    doc['Info'] = table.rows[1].cells[2].text.strip()
                    doc['Comment'] = table.rows[1].cells[3].text.strip()
                    

                    text = table.rows[2].cells[0].text.strip()
                    text = text.replace('C', 'С').replace('e', 'е').replace('o', 'о').replace('a', 'а')

                    doc['Text'] = text.strip()
                    doc['Collector'] = None
                    doc['Typewriter'] = None
                    doc['RawText'] = text
                    try:
                        collector = re.finditer('(Текст записан(.+?)|Сведения о собирателе: (.+?)|Собиратель: (.+?))\n|$', 
                                                text, re.DOTALL)
                        
#                         collector = re.finditer('Сведения о собирателе: (.+?)|Собиратель: (.+,)', text, re.DOTALL)
                        collector = list(collector)
                    except IndexError:
                        print('Wrong collector: ', filepath)
                        collector = None

                    try:
                        typewriter = re.finditer('(Набор текста: (.+?)|Наборщик: (.+?)|Сведения о наборщике текста: (.+?))(\n|$)', 
                                                 text, re.DOTALL)
#                         typewriter = re.finditer('Сведения о наборщике: (.+?)|', text, re.DOTALL)

                        typewriter = list(typewriter)
                    except IndexError:
                        print('Wrong typewriter: ', filepath)
                        typewriter = None
                    fl_collector = 0
                    fl_typewriter = 0
                    if any(collector):
                        for i, match in enumerate(collector[0].groups()[1:]):
                            if match is not None and not fl_collector:
                                doc['Collector'] = match.strip()
                                fl_collector = 1
                                doc['Text'] = text[:collector[0].start()].strip()
                            
                    if any(typewriter):
                        for i, match in enumerate(typewriter[0].groups()[1:]):
                            if match is not None and not fl_typewriter:
                                doc['Typewriter'] = match.strip()
                                fl_typewriter = 1
                    
                    
                    all_docs.append(doc)
```

```python
import json
json_str = json.dumps(all_docs)
with open('corpus.json', 'w', encoding='utf-8') as f:
    f.write(json_str)
```

```python
with open('corpus.json', 'r', encoding='utf-8') as f:
    corpus_json = json.load(f)
```
